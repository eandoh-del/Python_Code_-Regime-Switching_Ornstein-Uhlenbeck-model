### Monthly Aggregation and Missing Data Imputation
We will aggregate the daily data into monthly totals and then apply the combined average method:
$T_{month} = \frac{\sum_{t=1}^{7}(T-t)_m + \sum_{t=1}^{7}(T+t)_m}{14}$

```python
import pandas as pd
import numpy as np

# 1. Aggregate to Monthly Data
df_monthly = df_time_series.groupby(['Name', pd.Grouper(key='Date', freq='MS')])['Rainfall'].sum().reset_index()

# 2. Create a full monthly index for every station to identify missing months
all_stations = df_monthly['Name'].unique()
full_range = pd.date_range(start=df_monthly['Date'].min(), end=df_monthly['Date'].max(), freq='MS')

# Reindex to find gaps
mux = pd.MultiIndex.from_product([all_stations, full_range], names=['Name', 'Date'])
df_monthly_complete = df_monthly.set_index(['Name', 'Date']).reindex(mux).reset_index()

def impute_combined_average(group):
    """Applies the 7-period before/after average to NaNs"""
    data = group['Rainfall'].values
    n = len(data)
    imputed_data = data.copy()

    for i in range(n):
        if np.isnan(data[i]):
            # Indices for 7 periods before
            before = [i - t for t in range(1, 8) if i - t >= 0]
            # Indices for 7 periods after
            after = [i + t for t in range(1, 8) if i + t < n]

            combined_indices = before + after
            # Filter out indices that are also NaN
            valid_values = [data[idx] for idx in combined_indices if not np.isnan(data[idx])]

            if len(valid_values) > 0:
                imputed_data[i] = np.mean(valid_values)
            else:
                imputed_data[i] = 0 # Fallback if no data nearby exists

    group['Rainfall_Imputed'] = imputed_data
    return group

# Apply imputation per station
df_monthly_imputed = df_monthly_complete.groupby('Name', group_keys=False).apply(impute_combined_average)

# Check results for a station that had gaps
print("Monthly Imputation Summary:")
print(f"Original NaNs: {df_monthly_complete['Rainfall'].isna().sum()}")
print(f"Remaining NaNs after imputation: {df_monthly_imputed['Rainfall_Imputed'].isna().sum()}")

display(df_monthly_imputed.head())
```

## Hidden Markov Model (HMM) Analysis
Following the methodology in the research paper, we model cumulative seasonal rainfall ($R_t$) as a regime-switching Ornstein–Uhlenbeck (OU) process.

### Model Framework:
1.  **States ($S_t$):** Two unobserved climate regimes (Wet and Dry).
2.  **Dynamics:** $dR_t = \kappa_{S_t}(\theta_{S_t} - R_t)dt + \sigma_{S_t}dW_t$
3.  **Estimation:** Expectation-Maximization (EM) algorithm using Forward-Backward passes.

### HMM-OU with K-Means Initialization
We will now use the `sklearn.cluster.KMeans` library to initialize the regime parameters. This ensures that $\theta$ (long-run mean) and $\sigma$ (volatility) start at values that better represent the actual distribution of the rainfall data.

```python
import pandas as pd
import os

# Re-processing to explicitly exclude 2026 as requested
drive_path = '/content/drive/MyDrive/'
regions = {
    'Northern': 'Ghana_Northern_Pentadal.csv',
    'UpperWest': 'Ghana_Upper_West_Pentadal.csv'
}

regional_wide_filtered = {}

for name, file in regions.items():
    path = os.path.join(drive_path, file)
    if os.path.exists(path):
        df_reg = pd.read_csv(path)
        if 'Month' in df_reg.columns:
            df_reg['Month'] = df_reg['Month'].ffill()

        # Identify year columns (1981-2025) - Removing 2026
        year_cols = [str(y) for y in range(1981, 2026) if str(y) in df_reg.columns]
        id_cols = ['Month', 'Period']

        df_wide = df_reg[id_cols + year_cols].copy()
        regional_wide_filtered[name] = df_wide

        print(f"Processed {name} region (1981-2025).")
        print(f"Latest year in dataset: {year_cols[-1]}")
    else:
        print(f"Warning: {file} not found.")

# Update active variables
df_northern_wide = regional_wide_filtered.get('Northern')
df_upperwest_wide = regional_wide_filtered.get('UpperWest')

if df_northern_wide is not None:
    display(df_northern_wide.head())
```
